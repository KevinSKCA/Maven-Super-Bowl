```python
car_brand = ['Toyota', 'Hyundai', 'Kia']
dataset_edited_car = dataset_edited[dataset_edited.Brand.isin(car_brand)]
dataset_edited_car
```

```python
dataset_edited_car_two = dataset_edited_car.loc[dataset_edited_car.groupby('Year')['Youtube Views'].idxmax()].reset_index()
dataset_edited_car_two
```
